```python
import requests
import pandas as pd

# Create an empty list to store dataframes
dfs = []

for i in range(1, 429):
    response = requests.get(
        f'https://api.themoviedb.org/3/movie/top_rated?api_key=1dec709e4a302017193d0c7855ff1d45&language=en-US&page={i}'
    )
    
    # Check if the request succeeded
    if response.status_code == 200:
        results = response.json().get('results', [])
        if results:  # Only create a dataframe if results exist
            temp_df = pd.DataFrame(results)[
                ['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']
            ]
            dfs.append(temp_df)
    else:
        print(f"Failed to fetch page {i}: {response.status_code}")

# Combine all dataframes at once
df = pd.concat(dfs, ignore_index=True)
```
