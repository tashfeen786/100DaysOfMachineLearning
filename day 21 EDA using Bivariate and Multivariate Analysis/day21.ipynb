2. Bar Plot (Numerical - Catgorical)

```python
df.head()
```

```python
sns.barplot(data=df, x='pclass', y='age', hue='sex')
```

3. Box plot (Numerical - Catgorical)

```python
sns.boxplot(data= df, x='sex', y='age', hue='survived')
```

4. Displot (Numerical - Catgorical)

```python
sns.distplot(df[df['survived']==0]['age'],hist=False)
sns.distplot(df[df['survived']==1]['age'],hist=False)
```

5. Heatmap (Catgorical - Catgorical)

```python
df.head(3)
```

```python
pd.crosstab(df['pclass'], df['survived'])
```

```python
sns.heatmap(pd.crosstab(df['pclass'], df['survived']), annot=True, fmt='d') 
```

```python
(df.groupby('embarked')['survived'].mean()*100)
```

6. ClusterMap (Catgorical - Catgorical)
