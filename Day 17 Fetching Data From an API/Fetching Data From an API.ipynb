```python
import requests
import pandas as pd
import time

dfs = []
api_key = "1dec709e4a302017193d0c7855ff1d45"

for i in range(1, 429):
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=en-US&page={i}"
    
    for attempt in range(3):  # retry up to 3 times
        try:
            response = requests.get(url, timeout=10)
            response.raise_for_status()  # raise error for 4xx/5xx responses
            data = response.json().get("results", [])
            if data:
                temp_df = pd.DataFrame(data)[
                    ['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']
                ]
                dfs.append(temp_df)
            print(f"Page {i} done")
            time.sleep(0.3)  # small delay to avoid rate limits
            break  # success, move to next page
        except requests.exceptions.RequestException as e:
            print(f"Attempt {attempt+1} failed for page {i}: {e}")
            time.sleep(2)  # wait a bit before retrying
    else:
        print(f"Skipping page {i} after multiple failed attempts.")

# Combine all pages into one DataFrame
df = pd.concat(dfs, ignore_index=True)
```
